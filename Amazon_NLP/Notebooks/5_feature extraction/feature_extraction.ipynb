# Feature extraction

### feature extraction refers to the process of converting raw text data into a format that can be used for machine learning. Text data is inherently unstructured, and machine learning models require structured input. Feature extraction involves selecting and transforming relevant information from the raw text into a set of numerical features that can be used as input for a machine learning algorithm.

pseudo code:
function feature_extraction(raw_data):
    # Raw data is the input, such as a collection of text documents or images
    
    # Apply a specific feature extraction technique (e.g., TF-IDF for text data)
    features = apply_feature_extraction(raw_data)
    
    return features

```python
# Import necessary libraries
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.datasets import fetch_20newsgroups
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
# Load a sample dataset (you can replace this with your own dataset)
data = pd.read_csv('Sentimnal_file.csv')
data = data['Description_proc']
data.dropna(inplace = True)
```

```python
# Create a TF-IDF vectorizer
#using TfidfVectorizer() and adjust max_features to 5000
tfidf_vectorizer = TfidfVectorizer(max_features=5000)
```

```python
# Fit and transform the data
#using tfidf_vectorizer.fit_transform() to the data
tfidf_matrix = tfidf_vectorizer.fit_transform(data)
```

```python
# Get feature names
feature_names = np.array(tfidf_vectorizer.get_feature_names_out())
feature_names
```

```python
# Calculate mean TF-IDF score for each feature
mean_tfidf = tfidf_matrix.mean(axis=0).A1
```

```python
# Create a DataFrame for visualization
df_tfidf = pd.DataFrame({'Word': feature_names, 'Mean_TF-IDF': mean_tfidf})
```

```python
# Sort the DataFrame by Mean_TF-IDF in descending order
df_tfidf = df_tfidf.sort_values(by='Mean_TF-IDF', ascending=False)
```

```python
# Choose the top N words for visualization
top_n = 30
df_top_n = df_tfidf.head(top_n)
```

```python
# Plot the bar chart
plt.figure(figsize=(12, 8))
plt.barh(df_top_n['Word'], df_top_n['Mean_TF-IDF'], color='skyblue')
plt.xlabel('Mean TF-IDF Score')
plt.ylabel('Word')
plt.title(f'Top {top_n} Words based on TF-IDF')
plt.show()
```
