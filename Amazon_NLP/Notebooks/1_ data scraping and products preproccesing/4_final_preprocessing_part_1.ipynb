```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('general3')
```

### Read the data

```python
df.drop(['Unnamed: 0.1','Unnamed: 0'],axis=1,inplace=True)
```

```python
df
```

### clean the price column, then convert its type into integer 

```python
df['price'] = df['price'].str.replace(',','')
```

```python
df['price'] = df['price'].str.replace('.','')
df
```

```python
df['price'] = df['price'].str.replace(',', '', regex=True).str.replace('\D', '', regex=True).astype(int)
```

```python
df['price']
```

### making all the records in df['manifacture'] upper case 

#### as the same read can be defined into two or even more categorcategories 

```python
df['manifacture']=df['manifacture'].str.upper()
```

```python
df
```

### replace the null values in the discount with 0

#### as all null values doesnt have a discount

```python
df['discount']=df['discount'].fillna('0')
```

```python
df
```

### clean it then convert into int

```python
df['discount']=df['discount'].str.rstrip('%')
```

```python
df['discount']=df['discount'].str.replace('-','')
```

```python
df['discount']=df['discount'].str.replace('\D', '', regex=True).astype(int)
```

```python
df['discount']
```

### when scraping, when the product doesnt get any review, the website had a button instead called Previous page, so we'll erase it

### we will erase out of 5 stars out of all records that contain it

```python
df['rating']=df['rating'].str.replace('Previous page','')
df['rating']=df['rating'].str.replace('out of 5 stars','')
df
```

### we'll replace the space with null values right now untill we can deal with that later

```python
df['rating'].replace('', np.nan, inplace=True)
```

```python
df['rating'].astype(float)
```

```python
df['rating'].unique()
```

```python
df['rating'] = df['rating'].str.replace('5.0 من 5 نجوم','5.0')
```

```python
df['rating'].unique()
```

```python
df['rating']=df['rating'].astype(float)
```
