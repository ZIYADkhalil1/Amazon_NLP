```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
URL = 'https://www.amazon.eg/deal/912cee26?pf_rd_r=JTWKC8K5YSK39E4EDRJE&pf_rd_t=Events&pf_rd_i=blackfriday&pf_rd_p=ee55a2a2-1309-4ad7-bf63-ee00919ab1d3&pf_rd_s=slot-8&ref=dlx_black_gd_dcl_tlt_0_912cee26_dt_sl8_d3&pageNum=1&sortOption=price-asc-rank&language=en_AE'
```

```python
# Headers for request
HEADERS = ({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36', 'Accept-Language': 'en-US, en;q=0.5'}) #add your user agent 
```

```python
# HTTP Request
webpage = requests.get(URL, headers=HEADERS)
```

```python
# Soup Object containiang all data
soup = BeautifulSoup(webpage.content, "html.parser")
```

```python
soup
```

```python
# Fetch links as List of Tag Objects
links = soup.find_all("a", attrs={'class':'a-size-base a-color-base a-link-normal a-text-normal'})
```

```python
links
```

```python
links[0]
```

```python
link = links[0].get('href')
```

```python
product_list = "https://www.amazon.eg" + link
```

```python
new_webpage = requests.get(product_list, headers=HEADERS)
```
