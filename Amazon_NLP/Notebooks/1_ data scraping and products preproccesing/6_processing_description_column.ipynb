## we'll create a function that diffrent datasets according to grouping it by the department attribute with diffrent values

```python
def get_group(group):
    df1 = df.groupby('Department').get_group(group).reset_index()
    return df1
```

## this function cleans the description column

it iterates through each row, then iterates through the key-value pairs within each row. It checks whether each key-value pair has both a key and a value before updating the dictionary. The resulting DataFrame should be more robust to variations in the data.

```python
def get_description(df):
        # Initialize an empty list to store dictionaries
    attributes_list = []

    # Iterate through rows and update dictionaries
    for index, row in df.iterrows():
        attributes = {}
        for item in row['description'].split('     '):
            key_value = item.strip().split('   ')
            if len(key_value) == 2:
                key, value = key_value
                attributes[key] = value
        attributes_list.append(attributes)

    # Create a DataFrame from the list of dictionaries
    df_attributes = pd.DataFrame(attributes_list)

    # Concatenate the original DataFrame and the new DataFrame with attributes
    df = pd.concat([df, df_attributes], axis=1)

    # Display the structured DataFrame
    return df
```
